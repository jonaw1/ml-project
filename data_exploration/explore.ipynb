```python
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()

RAW_DATA_PATH = os.path.join('..', os.getenv("RAW_DATA_PATH")).replace('/', os.sep)

df = pd.read_csv(RAW_DATA_PATH)

display(df)
```

```python
for col in df.columns:
    if df[col].dtype != "int64" and df[col].dtype != "float64":
      print(col, df[col].unique(), df[col].dtype)
```

```python
nulls = df.isnull().sum()
nulls_filtered = nulls[nulls != 0]
nulls_filtered
```

```python
PRODUCTION_FOLDER = os.path.join('..', os.getenv("PRODUCTION_FOLDER")).replace('/', os.sep)

random = df.sample(20)
random = random.drop(columns=['Verkaufspreis'])
random.to_csv(os.path.join(PRODUCTION_FOLDER, 'test_data.csv'), index=False)
random
```
